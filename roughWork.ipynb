```python
#this is just for experimenting effect of log transform on the columns, the cat column will not be dropped 

df=data.drop(columns=['EJ','Id'], axis=1)
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import numpy as np

log_transformer=FunctionTransformer(np.log1p, inverse_func=np.expm1)
imputer=SimpleImputer(strategy='median')
scaler=StandardScaler()

num_pipeline=Pipeline([
    ('impute',imputer),
    ('log',log_transformer),
    ('standardize',scaler)
])

transformed_data=num_pipeline.fit_transform(df)
```

```python
transformed_df=pd.DataFrame(transformed_data, columns=df.columns)
transformed_df
```

```python
transformed_df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```
